```python
import os
import random
from shutil import copyfile

def train_val_split(source, train, val, train_ratio):
    os.makedirs(train, exist_ok=True)
    os.makedirs(val, exist_ok=True)

    total_size = len(os.listdir(source))
    train_size = int(total_size * train_ratio)

    randomized = random.sample(os.listdir(source), total_size)
    train_files = randomized[:train_size]
    val_files = randomized[train_size:]

    for i in train_files:
        source_file = os.path.join(source, i)
        destination = os.path.join(train, i)
        copyfile(source_file, destination)

    for i in val_files:
        source_file = os.path.join(source, i)
        destination = os.path.join(val, i)
        copyfile(source_file, destination)

train_ratio = 0.9


source_cats = "/content/dataset/archive/training_set/training_set/cats"
train_cats = "/content/dataset/archive/split/train/cats"
val_cats = "/content/dataset/archive/split/val/cats"

source_dogs = "/content/dataset/archive/training_set/training_set/dogs"
train_dogs = "/content/dataset/archive/split/train/dogs"
val_dogs = "/content/dataset/archive/split/val/dogs"

train_val_split(source_cats, train_cats, val_cats, train_ratio)
train_val_split(source_dogs, train_dogs, val_dogs, train_ratio)
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    horizontal_flip=True,
    shear_range=0.3,
    fill_mode='nearest',
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.1
)

test_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    horizontal_flip=True,
    shear_range=0.3,
    fill_mode='nearest',
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.1
)

validation_dir = "/content/dataset/test_set"
```

```python
val_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    horizontal_flip=True,
    shear_range=0.3,
    fill_mode='nearest',
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.1
)

validation_dir = "/content/dataset/archive/test_set/test_set"

val_generator = val_datagen.flow_from_directory(
    validation_dir,
    target_size=(150, 150),
    batch_size=10,
    class_mode='binary'
)

print("Validasi Generator Berhasil Dibuat ✅")
```

```python
train_generator = train_datagen.flow_from_directory(
    "/content/dataset/archive/training_set/training_set",
    target_size=(150, 150),
    batch_size=10,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    "/content/dataset/archive/test_set/test_set",
    target_size=(150, 150),
    batch_size=10,
    class_mode='binary'
)
```

```python
import tensorflow as tf

class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        acc = logs.get('accuracy') or logs.get('acc')
        if acc is not None and acc > 0.99:
            print(f"\nAkurasi {acc:.2%} mencapai 99%, training dihentikan.")
            self.model.stop_training = True

callbacks = myCallback()
```

```python
import tensorflow as tf

tf.random.set_seed(112)

model = tf.keras.models.Sequential([

    tf.keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),


    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),


    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),


    tf.keras.layers.Flatten(),


    tf.keras.layers.Dense(200, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),


    tf.keras.layers.Dense(500, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.5),


    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='Adam',
    metrics=['accuracy']
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping


history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator) // 2,
    epochs=50,
    validation_data=val_generator,
    validation_steps=len(val_generator),
    verbose=1,
)
```

```python
%matplotlib inline

import matplotlib.image as mpimg
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.legend(loc='best')
plt.figure()
plt.show()

plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and Validation Loss')
plt.legend(loc='best')
plt.figure()
plt.show()
```

```python
import numpy as np
from keras.preprocessing import image
from google.colab import files

uploaded = files.upload()

class_list = ["cats", "dogs"]

for fn in uploaded.keys():

    path = fn
    img = image.load_img(path, target_size=(150, 150))
    plt.imshow(img)
    plt.axis('off')
    plt.show()

    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = x / 255.0

    classes = model.predict(x)

    if classes.shape[1] == 2:
        predicted_class = np.argmax(classes, axis=1)[0]
    else:
        predicted_class = int(classes[0][0] > 0.5)

    print(f"Prediksi: This is a {class_list[predicted_class]}")
```
